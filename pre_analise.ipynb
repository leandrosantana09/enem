```python
import pandas as pd
from info_apoio import missing_values_table
```

# 1 - LEITURA E ANÁLISE DOS DADOS

```python
base = pd.read_csv("DADOS\MICRODADOS_ENEM_2021.csv",
                   encoding="latin1", 
                   sep=';'
                   )
base.head()
```

```python
base = base.loc[base['SG_UF_ESC']=='MG'] # apenas MG
base.head(5)
```

```python
colunas = base.columns
colunas
```

```python
# 
```

```python
only_one = []
print("Coluna e quantidade de itens diferentes\n")
for col in colunas:
    teste = list(base[col].value_counts())
    size = len(teste)
    print(col, ':',size)
    if size == 1:
      only_one.append(col)


only_one
```

```python
many_values = ['TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_LC',
               'TX_RESPOSTAS_MT', 'TP_LINGUA', 'TX_GABARITO_CN', 'TX_GABARITO_CH',
               'TX_GABARITO_LC', 'TX_GABARITO_MT', 'TP_STATUS_REDACAO', 
               'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC']
```

```python
lista_drop = only_one + many_values

base.drop(columns=lista_drop,
          inplace=True
          )

base.head(5)
```

```python
missing_values_table(base) # missing_values_table retorna todas as colunas com valores nulos, assim como a porcentagem referente
```
