```python
import pandas as pd
import numpy as np

from info_apoio import best_corr
from info_apoio import columns_null
from info_apoio import drop_outlier
```

# Lendo os dados

```python
base = pd.read_csv("DADOS\dado_mg.csv", 
                   sep=';'
                   )
base.head()
```

```python
base_number = base.select_dtypes('number') # Selecionando apenas as colunas com dados do tipo 'number'
base_number
```

Quantidade de valores diferentes por partidas

```python
for x in base_number.columns:
    print(x, len(base_number[x].value_counts()))
```

Criando uma lista com colunas de valor único

```python
colunas_unicas = []
for x in base_number.columns:
    if len(base_number[x].value_counts()) == 1:
        print(x, len(base_number[x].value_counts()))
        colunas_unicas.append(x)              # lista contendo as colunas com valores unicos
```

Retirando colunas com valores únicos

```python
base_number.drop(columns = colunas_unicas, inplace=True)
base_number
```

```python
len(base_number)
```

Quantidade de valores nulos das colunas.

```python
base_number.isnull().sum().sort_values(ascending=False).head(10)
```

Selecionando as colunas que tem maior porcentagem de valores nulos.

```python
missing_values_columns = columns_null(base_number, 6)
missing_values_columns
```

Removendo colunas

```python
base_number.drop(columns=missing_values_columns, # removendo as colunas com mais valores nulos
                 inplace=True
                 )
base_number
```

# Calculando a MÉDIA PONDERADA

```python
base_nota = ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'NU_NOTA_LC']
base_number[base_nota]
```

```python
base_number['MEDIA'] = (base_number[base_nota]['NU_NOTA_CN'] +
                 3*base_number[base_nota]['NU_NOTA_CH'] +
                 base_number[base_nota]['NU_NOTA_MT'] +
                 3*base_number[base_nota]['NU_NOTA_REDACAO'] +
                 2*base_number[base_nota]['NU_NOTA_LC'])/10
```

```python
base_nota.append('MEDIA')
base_number[base_nota]
```

```python
base_number[base_nota].isna().sum()
```

Preenchendo os valores nulos com a média de cada conteúdo 

```python
base_number['NU_NOTA_CH'].fillna(base_number['NU_NOTA_CH'].mean(), inplace=True)
base_number['NU_NOTA_CN'].fillna(base_number['NU_NOTA_CN'].mean(), inplace=True)
base_number['NU_NOTA_MT'].fillna(base_number['NU_NOTA_MT'].mean(), inplace=True)
base_number['NU_NOTA_REDACAO'].fillna(base_number['NU_NOTA_REDACAO'].mean(), inplace=True)
base_number['NU_NOTA_LC'].fillna(base_number['NU_NOTA_LC'].mean(), inplace=True)
```

```python
base_number[base_nota]
```

```python
base_number['MEDIA_NOVA'] = (base_number[base_nota]['NU_NOTA_CN'] +
                 3*base_number[base_nota]['NU_NOTA_CH'] +
                 base_number[base_nota]['NU_NOTA_MT'] +
                 3*base_number[base_nota]['NU_NOTA_REDACAO'] +
                 2*base_number[base_nota]['NU_NOTA_LC'])/10
base_nota.append('MEDIA_NOVA')
```

```python
base_number[base_nota]
```

```python
base_number[base_nota].describe()
```

```python
base_number.corr()['MEDIA_NOVA'].sort_values(ascending=False)
```

```python
base_number.drop(columns=['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_REDACAO', 'NU_NOTA_LC', 
                          'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4',
                          'NU_NOTA_COMP5', 'MEDIA'], inplace=True)
```

Vamos selecionar as 60 colunas com maior modulo de correlação.

```python
lista_top_columns = best_corr(base_number, 'MEDIA_NOVA', 30, 30)
lista_top_columns
    
```

```python
base_number = base_number[lista_top_columns]
```
