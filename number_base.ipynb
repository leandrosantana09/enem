```python
import pandas as pd
import numpy as np
```

# 1 - LENDO OS DADOS

```python
base = pd.read_csv('base_mg_number.csv')
base.head()
```

# 2 - ANÁLISE DAS NOTAS FALTANTES E CALCULO DA MÉDIA

```python
FREQ_CH = np.where(base.NU_NOTA_CH.isnull(),1,0)
FREQ_CN = np.where(base.NU_NOTA_CN.isnull(),1,0)
FREQ_MT = np.where(base.NU_NOTA_MT.isnull(),1,0)
FREQ_LC = np.where(base.NU_NOTA_LC.isnull(),1,0)
FREQ_REDACAO= np.where(base.NU_NOTA_REDACAO.isnull(),1,0)
```

```python
FREQ_SUM = FREQ_CH + FREQ_CN + FREQ_MT + FREQ_REDACAO + FREQ_LC
base['TOTAL_NAN'] = FREQ_SUM
```

```python
base['TOTAL_NAN'].value_counts()
```

```python
base = base.loc[(base['TOTAL_NAN'] != 3) & (base['TOTAL_NAN'] != 5)]
base
```

```python
# substituindo dados faltantes pela média

base['NU_NOTA_CH'].fillna(base['NU_NOTA_CH'].mean(), inplace=True)
base['NU_NOTA_CN'].fillna(base['NU_NOTA_CN'].mean(), inplace=True)
base['NU_NOTA_MT'].fillna(base['NU_NOTA_MT'].mean(), inplace=True)
base['NU_NOTA_REDACAO'].fillna(base['NU_NOTA_REDACAO'].mean(), inplace=True)
base['NU_NOTA_LC'].fillna(base['NU_NOTA_LC'].mean(), inplace=True)
```

```python
# calculo da média

base['MEDIA'] = (base['NU_NOTA_CN'] +
                 base['NU_NOTA_CH'] +
                 base['NU_NOTA_MT'] +
                 base['NU_NOTA_REDACAO'] +
                 base['NU_NOTA_LC'])/5
```

```python
base.head()
```

```python
# remoção das colunas que não vão ser mais usadas

base.drop(columns=['TOTAL_NAN',
                   'NU_NOTA_CN',
                   'NU_NOTA_CH',
                   'NU_NOTA_LC',
                   'NU_NOTA_MT',
                   'NU_NOTA_REDACAO'], inplace=True)

base
```

# 3 - SALVANDO OS DADOS E ENSERANDO ESTÁ PARTE DA ANÁLISE

```python
base.to_csv('DADOS_TRATADOS\data_number_final.csv',
            index=False)
```
