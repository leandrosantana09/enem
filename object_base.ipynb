```python
import pandas as pd
import numpy as np

from info_apoio import columns_null
```

# Lendo os dados:

```python
base = pd.read_csv("DADOS\dado_mg.csv", 
                   sep=';'
                   )
base.head()
```

```python
base_object =  base.select_dtypes('object') # Selecionando apenas as colunas com dados do tipo 'object'
```

Da mesma maneira usada no caso numerico, a função abaixo ira retornar a coluna e a quantidade de valores diferentes da mesma.

```python
for x in base_object.columns:
    print(x, len(base_object[x].value_counts()))
```

Diferente do caso numerico, além de selecionar apenas as colunas com valores unicos, também vou escolher as que contenham muitos valores diferentes.

=> As colunas 'TX_GABARITO_CN, TX_GABARITO_CH, TX_GABARITO_LC, TX_GABARITO_MT' também foram deletadas, pois não apresentam interesse para tal analise.

```python
colunas_drop = ['NO_MUNICIPIO_RESIDENCIA', 'NO_MUNICIPIO_NASCIMENTO', 'NO_MUNICIPIO_ESC', 'NO_MUNICIPIO_PROVA',
                 'TX_RESPOSTAS_CN', 'TX_RESPOSTAS_CH', 'TX_RESPOSTAS_CN', 'TX_RESPOSTAS_LC', 'TX_RESPOSTAS_MT',
                 'TX_GABARITO_CN', 'TX_GABARITO_CH', 'TX_GABARITO_LC', 'TX_GABARITO_MT', 'SG_UF_PROVA']
```

```python
base_object.drop(columns = colunas_drop, # Removendo as colunas da lista 'colunas_drop'
                 inplace=True            # Salvando as alterações feitas 
                 )
base_object
```

Agora vamos analisar os dados faltantes.

```python
base_object.isnull().sum().sort_values(ascending=False).head(10)
```

```python
len(base_object) # Total de linhas do dataset
```

As colunas 'SG_UF_ESC' e 'SG_UF_NASCIMENTO' se destacam como sendo as unicas com valores faltantes, esses valores que correspondem de forma significativa ao total de dados, assim sendo vou removelos.

```python
'''A função columns_null retorna uma lista com as n colunas com mais valores ausentes do datset escolhido'''
missing_values_columns = columns_null(base_object,
                                      n=2)
missing_values_columns
```

```python
base.drop(columns=missing_values_columns, # Removendo as colunas da lista 'missing_values_columns'
          inplace=True                    # Salvando as alterações feitas 
          )
base
```

```python
base.head(10)
```

Como forma alternativa de trabalhar com variaveis categoricas, vou usar o metodo 'get_dummies' que faz a converção de variaveis categoricas em indicadores.

```python
base_object = pd.get_dummies(data=base_object)
base_object
```

```python
base_object.insert(0, 'NU_INSCRICAO',                # Add a coluna 'NU_INSCRICAO', 
                   base['NU_INSCRICAO'].to_list(),   # sera usada como chave para a união dos dados no arquivo 'base_analise'
                   True
                   )
base_object
```

# Salvando os dados e encerrando esse parte da análise. 

```python
base_object.to_csv('data_object.csv',
                   index=False)
```
