```python
import pandas as pd
import numpy as np

from info_apoio import columns_null
```

# 1 - LENDO OS DADOS

```python
base = pd.read_csv('base_mg_object.csv')
base.head()
```

```python
base.columns
```

```python
# colunas sem interferencia significativa

colunas_drop = ['TP_DEPENDENCIA_ADM_ESC', 'TP_LOCALIZACAO_ESC', 'TP_SIT_FUNC_ESC',
                'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA', 'CO_UF_PROVA',
                'SG_UF_PROVA', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC',
                'TP_PRESENCA_MT', 'CO_PROVA_CN', 'CO_PROVA_CH', 'CO_PROVA_LC',
                'CO_PROVA_MT', 'TP_FAIXA_ETARIA']
```

```python
base.drop(columns = colunas_drop, # Removendo as colunas da lista 'colunas_drop'
          inplace=True            # Salvando as alterações feitas 
          )
base
```

# 2 - ANÁLISE DOS DADOS FALTANTES 

```python
base.fillna(base.mode().iloc[0],
                   inplace=True
                   )
base
```

```python
base.info()
```

```python
# conversão das colunas para object

base['TP_ESTADO_CIVIL'] = base['TP_ESTADO_CIVIL'].astype(object)
base['TP_COR_RACA'] = base['TP_COR_RACA'].astype(object)
base['TP_NACIONALIDADE'] = base['TP_NACIONALIDADE'].astype(object)
base['TP_ESCOLA'] = base['TP_ESCOLA'].astype(object)
base['TP_ENSINO'] = base['TP_ENSINO'].astype(object)
base['Q005'] = base['Q005'].astype(object)
```

```python
base.info()
```

```python
base.to_csv('analise_grafica.csv',
            index=False)
```

# 3 - CONVERSÃO DE VARIAVEIS

Como forma alternativa de trabalhar com variaveis categoricas, vou usar o metodo 'get_dummies' que faz a converção de variaveis categoricas em indicadores.

```python
base = pd.get_dummies(data=base)
base.head(5)
```
